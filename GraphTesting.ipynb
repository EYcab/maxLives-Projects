```python
import pandas as pd
from pandas import Series,DataFrame
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

test_db = pd.read_csv('C:\\Users\\Chenxi\\Desktop\\python\\Projects\\testStar.csv')
test_Multiple_Peaks_db = pd.read_csv('C:\\Users\\Chenxi\\Desktop\\python\\Projects\\testMultiplePeaks.csv')
```

```python
data_len=len(test_db.axes[0])
BirthYr = test_db['Birth-Year'].dropna()
DeathYr = test_db['Death-Year'].dropna()
#print live_period
livesInYr = []
for j in range(0,data_len):
    for yr in range(BirthYr[j],DeathYr[j]+1):
        livesInYr.append(yr)
DataFrame(livesInYr).hist(figsize=(16,10),bins=100,xlabelsize=15)
```

```python
data_len=len(test_Multiple_Peaks_db.axes[0])
BirthYr = test_Multiple_Peaks_db['Birth-Year'].dropna()
DeathYr = test_Multiple_Peaks_db['Death-Year'].dropna()
#print live_period
livesInYr = []
for j in range(0,data_len):
    for yr in range(BirthYr[j],DeathYr[j]+1):
        livesInYr.append(yr)
DataFrame(livesInYr).hist(figsize=(16,10),bins=100,xlabelsize=15)

```
